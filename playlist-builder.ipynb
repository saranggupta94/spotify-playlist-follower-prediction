```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.options.display.max_columns = None
```

```python
data = pd.read_csv('datasets/allTracksInfo.csv')
data.drop(columns="Unnamed: 0", inplace=True)
data.head()
```

```python
dataTracks = data[['name', 'id', 'artist_name', 'acousticness', 'danceability', 'duration_ms', 'energy', 'explicit', 'instrumentalness',
     'key', 'liveness', 'mode', 'popularity', 'speechiness', 'tempo', 'time_signature', 'valence']]
```

```python
dataTracks.drop_duplicates(inplace=True)
```

```python
dataTracks = pd.get_dummies(dataTracks, columns = ['time_signature', 'key'], prefix=['time_sign', 'key'])
dataTracks['explicit'] = dataTracks['explicit']*1
dataTracks.dropna(inplace=True)
```

```python
dataTracks.set_index('id', inplace=True)
dataTracks.head()
```

```python
from sklearn.metrics.pairwise import euclidean_distances

def createPlaylist(track, dataTracks, length):
    dataTracks['distance'] = euclidean_distances(pd.DataFrame(dataTracks.drop(columns=['name', 'artist_name']).loc[track]).values.reshape(1,-1), 
                                                        dataTracks.drop(columns=['name', 'artist_name']))[0]
    return dataTracks.loc[dataTracks['distance'].sort_values().head(length).index][['name', 'artist_name']]
```

```python
#Specify track id, dataframe and number of tracks in the playlist
createPlaylist('4c6PicLrAOwfkA7yEMwmI5', dataTracks, 20)
```
